<h2 style="margin-top: 30px; margin-bottom: 15px; color: #2c3e50;">1) Descrizione del problema presentato</h2>
<p style="text-align: normalize; margin-bottom: 15px;">In questo progetto si studia come costruire portafogli azionari che rispettino vincoli pratici di gestione: numero massimo di titoli in portafoglio, presenza minima di emittenti considerati “affidabili” e limiti sui pesi di ogni titolo. L’idea di base è combinare il classico modello media-varianza con strumenti presi dalla teoria dei grafi.<sup><a href="#ref-4">[4]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">L’universo iniziale è formato da titoli dell’S&P 500. A partire da questo insieme si costruiscono portafogli coerenti con il profilo di rischio dell’investitore, cercando di ottenere una buona diversificazione rispetto alle correlazioni storiche tra i rendimenti e monitorando la qualità del credito delle imprese, distinguendo tra titoli con rating e titoli senza rating.<sup><a href="#ref-10">[10]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Nel lavoro sono state identificate due famiglie principali di soluzioni basate sul concetto di “portafoglio su grafo”:</p>
<ol style="margin-bottom: 15px; padding-left: 40px;">
<li style="text-align: justify; margin-bottom: 5px;">un portafoglio a minima varianza con vincoli realistici sui pesi e sulla cardinalità;<sup><a href="#ref-6">[6]</a></sup></li>
<li style="text-align: justify; margin-bottom: 5px;">portafogli “decorrelati” costruiti usando il grafo delle distanze tra i titoli.<sup><a href="#ref-3">[3]</a></sup></li>
</ol>
<p style="text-align: justify; margin-bottom: 15px;">La gestione dei titoli privi di rating è emersa come uno dei punti più delicati. La variabile “unrated” non rimane una semplice etichetta descrittiva, ma diventa un vero vincolo quantitativo da rispettare nella costruzione del portafoglio.<sup><a href="#ref-6">[6]</a></sup></p>
<h3 style="margin-top: 25px; margin-bottom: 10px; color: #34495e;">1.1 Descrizione dell’importanza gestionale del problema</h3>
<p style="text-align: justify; margin-bottom: 15px;">Dal punto di vista gestionale, il progetto propone uno strumento che non si limita a ridurre la varianza del portafoglio, ma integra anche vincoli sulla qualità media del portafoglio (rating), requisiti operativi (cardinalità K) e controllo della correlazione.<sup><a href="#ref-6">[6]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">In questo modo è possibile ridurre la concentrazione, limitare l’esposizione a titoli poco trasparenti e adattare la quota di titoli “unrated” al profilo di rischio scelto. Il sistema produce due portafogli da confrontare: un portafoglio a minima varianza e un portafoglio a massima decorrelazione.<sup><a href="#ref-4">[4]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">In una fase iniziale la teoria dei grafi è stata usata soprattutto come strumento esplorativo, per analizzare cluster, centralità e comunità nel network dei titoli. Con il procedere del lavoro, però, l’uso più utile dal punto di vista gestionale si è rivelato quello del grafo come filtro per selezionare l’universo dei titoli e come base per costruire portafogli con bassa correlazione interna.<sup><a href="#ref-3">[3]</a></sup></p>
<h3 style="margin-top: 25px; margin-bottom: 10px; color: #34495e;">1.2 Descrizione iniziale degli algoritmi coinvolti</h3>
<p style="text-align: justify; margin-bottom: 15px;">Il progetto integra più blocchi algoritmici all’interno di un unico modello (Model) e delle relative classi di supporto (DAO, RiskEstimator, GraphBuilder, PortfolioSelector, PortfolioWeights).<sup><a href="#ref-2">[2]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">1.2.1 Costruzione di correlazioni e distanze</h4>
<p style="text-align: justify; margin-bottom: 15px;">Il flusso di base per passare dai dati grezzi al grafo prevede il calcolo dei rendimenti logaritmici, la stima della matrice di covarianza $\Sigma$ e la definizione di una matrice delle distanze $D = (d_{ij})$, tipicamente definita come $d_{ij} = 1 - |\rho_{ij}|$. In questo modo il livello statistico è collegato in modo diretto al livello di rete all’interno di un flusso coerente.<sup><a href="#ref-3">[3]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">1.2.2 Filtraggio del grafo e definizione dell’universo ristretto U'</h4>
<p style="text-align: justify; margin-bottom: 15px;">A partire da $D$, la classe GraphBuilder costruisce un grafo pesato. Sul grafo grezzo si applicano filtri per ridurre rumore e densità, come soglie sulla correlazione e filtri k-NN che mantengono solo i $k$ vicini più simili.<sup><a href="#ref-5">[5]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Invece di usare algoritmi come BFS/DFS per generare direttamente i portafogli a partire dai cluster, si è scelto un approccio in due step: usare il grafo come filtro strutturale per definire un universo compatto $U'$ e delegare la selezione combinatoria a una procedura ricorsiva ottimizzata.<sup><a href="#ref-4">[4]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">1.2.3 Gestione del mix rated/unrated</h4>
<p style="text-align: justify; margin-bottom: 15px;">Ogni titolo viene classificato come “rated” o “unrated”. Per ogni profilo di rischio sono impostati parametri come la quota massima di peso assegnabile a titoli non classificati (`max_unrated_share`) e il valore minimo del rating medio. Questi vincoli vengono controllati sia in fase di selezione che in fase di calcolo dei pesi.<sup><a href="#ref-10">[10]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">1.2.4 Vincolo di peso minimo 1/(2K)</h4>
<p style="text-align: justify; margin-bottom: 15px;">Al modello media-varianza classico sono stati aggiunti vincoli operativi, tra cui il peso minimo per ogni titolo selezionato pari a $w_{\min} = \frac{1}{2K}$. Senza questo limite, il modello tende a concentrare gran parte del capitale su pochi titoli; imporre questo vincolo obbliga invece a una diversificazione minima tra i titoli scelti.<sup><a href="#ref-6">[6]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">1.2.5 Ricorsione per la selezione combinatoria</h4>
<p style="text-align: justify; margin-bottom: 15px;">La scelta di $K$ titoli sotto vincoli è un problema combinatorio affrontato con una procedura ricorsiva (Branch & Bound). Per ridurre i tempi di calcolo sono state introdotte ottimizzazioni come pre-filtri quantitativi e l’uso di NumPy per le parti numeriche più pesanti, limitando l’overhead di Python puro.<sup><a href="#ref-11">[11]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">1.2.6 Algoritmo di Dijkstra per portafogli decorrelati</h4>
<p style="text-align: justify; margin-bottom: 15px;">Sul grafo delle distanze si applica l’algoritmo di Dijkstra per calcolare i percorsi minimi. L’obiettivo è selezionare titoli che si trovano su percorsi caratterizzati da archi con distanza elevata (bassa correlazione).<sup><a href="#ref-9">[9]</a></sup> Su questo insieme ridotto si calcolano poi i pesi ottimali mantenendo i vincoli sul mix rated/unrated.<sup><a href="#ref-6">[6]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">1.2.7 Simulazione Monte Carlo</h4>
<p style="text-align: justify; margin-bottom: 15px;">Per ogni combinazione di portafoglio è stato implementato un modulo di simulazione Monte Carlo che genera scenari su orizzonti di 10, 20 e 30 anni. I risultati vengono riassunti tramite grafici delle distribuzioni, valori attesi e quantili di rischio.<sup><a href="#ref-8">[8]</a></sup></p>
<h3 style="margin-top: 25px; margin-bottom: 10px; color: #34495e;">1.3 Descrizione iniziale delle funzionalità software</h3>
<p style="text-align: justify; margin-bottom: 15px;">L’applicazione è stata sviluppata come interfaccia desktop in Python, seguendo un’architettura Model–View–Controller.<sup><a href="#ref-2">[2]</a></sup> L’utente inserisce capitale e numero di titoli ($K$), sceglie il profilo di rischio e il software costruisce automaticamente i portafogli (minima varianza e decorrelato) rispettando i vincoli sui rating.<sup><a href="#ref-10">[10]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">L’applicazione gestisce anche i casi in cui i vincoli sono troppo restrittivi, mostrando messaggi di errore che invitano a rilassare i parametri. È presente un sistema di log per monitorare i tempi di calcolo.<sup><a href="#ref-11">[11]</a></sup></p>

<h2 style="margin-top: 30px; margin-bottom: 15px; color: #2c3e50;">2. Descrizione del problema trattato</h2>
<h3 style="margin-top: 25px; margin-bottom: 10px; color: #34495e;">2.1 Contesto operativo e aziendale</h3>
<p style="text-align: justify; margin-bottom: 15px;">Nel modello di <b>Markowitz</b>, un portafoglio definito dai pesi $w$ (tali che $\sum_i w_i=1$) è caratterizzato da un rendimento atteso $m^{\top} w$ e da una varianza $w^{\top}\Sigma w$. Il problema del portafoglio a <b>minima varianza</b> si formula come segue:<sup><a href="#ref-1">[1]</a></sup></p>
<div style="text-align: center; margin: 15px 0;">
$$ \min_{w} \frac{1}{2} w^{\top} \Sigma w \quad s.t. \quad m^{\top} w \geq \mu_b, \quad e^{\top} w = 1 $$
</div>
<p style="text-align: justify; margin-bottom: 15px;">Dove $e$ è il vettore unitario e $\mu_b$ rappresenta il rendimento minimo accettabile. Poiché la matrice di covarianza $\Sigma$ e il vettore dei rendimenti attesi $m$ non sono osservabili, è necessario stimarli. Tuttavia, la covarianza campionaria può essere disturbata da rumore e collinearità. Per ovviare a questo problema, si utilizza uno <b>stimatore shrinkage</b>: la matrice stimata si ottiene come una combinazione convessa tra la matrice empirica $\widehat{\Sigma}$ e una matrice target $T$ (ad esempio diagonale). Il parametro di shrinkage $\lambda\in[0,1]$ regola quanto peso dare alla struttura campionaria.<sup><a href="#ref-1">[1]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Nel contesto di questo lavoro, l'analisi si basa su un universo di centinaia di azioni statunitensi, arricchito da un dataset di <b>rating creditizi</b>. Ogni titolo possiede un rating alfabetico (da AAA a D), che viene convertito in un punteggio numerico su scala 1-22. Le definizioni distinguono tra titoli <b>investment grade</b> (es. AAA) e titoli <b>speculativi</b> (BB/B) o prossimi al default. Oltre a ottimizzare rischio e rendimento, la costruzione del portafoglio deve rispettare vincoli specifici: diversificazione settoriale, limiti sulla quota di titoli senza rating (<i>unrated</i>) e soglie per il rating medio.<sup><a href="#ref-10">[10]</a></sup></p>
<h3 style="margin-top: 25px; margin-bottom: 10px; color: #34495e;">2.2 Sotto-problema selezionato e flusso logico</h3>
<p style="text-align: justify; margin-bottom: 15px;">Nel lavoro ci concentriamo su un sotto-problema concreto: data una base dati ampia, vogliamo selezionare un portafoglio di dimensione fissa ($K$) che massimizzi uno <b>score complessivo</b>, rispettando una serie di vincoli.<sup><a href="#ref-6">[6]</a></sup></p>
<p style="text-align: justify; margin-bottom: 10px;">I vincoli principali sono:</p>
<ul style="margin-bottom: 15px; padding-left: 40px;">
<li style="text-align: justify; margin-bottom: 5px;">una percentuale minima di titoli con rating elevato (≥ <code>rating_min</code>);</li>
<li style="text-align: justify; margin-bottom: 5px;">una percentuale massima di titoli senza rating (<code>max_unrated_share</code>);</li>
<li style="text-align: justify; margin-bottom: 5px;">un limite alla correlazione assoluta tra coppie di titoli ($|\rho_{ij}| \leq \rho_{\text{pair\_max}}$);</li>
<li style="text-align: justify; margin-bottom: 5px;">una quota massima per ogni settore (<code>max_share_per_sector</code>);</li>
<li style="text-align: justify; margin-bottom: 5px;">un peso minimo per ciascun titolo ($w_{\min} = 1/(2K)$).<sup><a href="#ref-7">[7]</a></sup></li>
</ul>
<p style="text-align: justify; margin-bottom: 15px;">Per arrivare al portafoglio finale, il codice segue un flusso di passi ben definiti.</p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">1. Caricamento dei dati e modellazione degli asset</h4>
<p style="text-align: justify; margin-bottom: 15px;">Il modulo <code>DAO</code> legge i file CSV e, per ogni ticker, costruisce un oggetto <code>Stock</code> che contiene serie storiche dei prezzi, rendimenti logaritmici, punteggio di rating, settore e flag di classificazione. In questo modo la logica di caricamento rimane isolata e il resto del modello lavora su oggetti già “preparati”.<sup><a href="#ref-11">[11]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">2. Valutazione dei parametri di rischio</h4>
<p style="text-align: justify; margin-bottom: 15px;">A questo punto i dati grezzi vengono trasformati in input per il modello:</p>
<ul style="margin-bottom: 15px; padding-left: 40px;">
<li style="text-align: justify; margin-bottom: 5px;"><b>Pulizia dei dati:</b> Si calcolano i rendimenti logaritmici giornalieri e si applica la <i>winsorizzazione</i> per ridurre l'impatto dei valori estremi. Le serie con troppi valori mancanti vengono scartate.<sup><a href="#ref-11">[11]</a></sup></li>
<li style="text-align: justify; margin-bottom: 5px;"><b>Stima della covarianza:</b> Si calcola la covarianza empirica ($\widehat{\Sigma}$), poi “corretta” con lo shrinkage ($\Sigma_{\text{sh}}$) per migliorare la stabilità numerica.<sup><a href="#ref-1">[1]</a></sup></li>
<li style="text-align: justify; margin-bottom: 5px;"><b>Correlazioni:</b> Da $\Sigma_{\text{sh}}$ si ricava la matrice di correlazione $\rho$ e si stimano i rendimenti attesi $\mu$.<sup><a href="#ref-3">[3]</a></sup></li>
</ul>
<h4 style="margin-top: 20px; margin-bottom: 5px;">3. Costruzione del grafo di correlazione e universo ridotto (U')</h4>
<p style="text-align: justify; margin-bottom: 15px;">Per analizzare le relazioni tra titoli usiamo un <b>grafo</b>. A partire dalla matrice di correlazione, costruiamo una matrice delle distanze $D$ (tipicamente $d_{ij} = 1 - |\rho_{ij}|$). Su questa matrice si applicano un filtro a soglia (se $|\rho_{ij}| < \tau$ il collegamento è eliminato) e un filtro k-NN che mantiene solo i $k$ legami più forti.<sup><a href="#ref-3">[3]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">In parallelo, si costruisce l’universo ridotto $U'$ filtrando i titoli in base a rating minimo, quota di <i>unrated</i> e dimensione massima. L’insieme $U'$ è l’universo investibile “pulito” su cui lavorano gli algoritmi combinatori.<sup><a href="#ref-6">[6]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">4. Selezione combinatoria (Branch & Bound)</h4>
<p style="text-align: justify; margin-bottom: 15px;">Dall’universo $U'$ vogliamo estrarre $K$ titoli che massimizzino uno score complessivo. Il problema è combinatorio e viene gestito con un algoritmo <b>Branch & Bound</b>. Si parte da una pre-selezione basata su score individuali, per poi esplorare i rami promettenti verificando i vincoli su pesi minimi e rating medio.<sup><a href="#ref-6">[6]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">La funzione obiettivo per un portafoglio $S$ considera correlazione media, rating, penalità settoriale e rendimento atteso, pesati dai parametri $\alpha, \beta, \gamma, \delta$.<sup><a href="#ref-7">[7]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">5. Calcolo dei pesi</h4>
<p style="text-align: justify; margin-bottom: 15px;">Una volta scelti i $K$ titoli, si decide quanto investire su ciascuno. Le logiche adottate sono <b>Equal-weight</b> (pesi uguali) oppure <b>Mean-variance</b> (problema di Markowitz con vincoli). Per evitare problemi numerici si introduce una regolarizzazione sulla matrice di covarianza.<sup><a href="#ref-2">[2]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">6. Portafoglio de-correlato tramite Dijkstra</h4>
<p style="text-align: justify; margin-bottom: 15px;">Accanto al portafoglio combinatorio, si costruisce un portafoglio alternativo usando l’algoritmo di <b>Dijkstra</b> sul grafo delle distanze. L’idea è selezionare titoli che siano “lontani” (meno correlati) partendo da un nodo sorgente e aggiungendo progressivamente i nodi più distanti fino a raggiungere $K$ nomi.<sup><a href="#ref-9">[9]</a></sup></p>
<h4 style="margin-top: 20px; margin-bottom: 5px;">7. Simulazioni Monte Carlo</h4>
<p style="text-align: justify; margin-bottom: 15px;">Per valutare il comportamento futuro si usano simulazioni <b>Monte Carlo</b>, assumendo un processo log-normale. Si generano `n_paths` scenari su un orizzonte temporale definito, analizzando statistiche come il valore atteso finale, i quantili (es. 5% e 95%) e la probabilità di perdita.<sup><a href="#ref-8">[8]</a></sup></p>
<h3 style="margin-top: 25px; margin-bottom: 10px; color: #34495e;">2.3 Principali input</h3>
<p style="text-align: justify; margin-bottom: 15px;">Il modello richiede input di tipo “dati” (prezzi, rating settoriali) e parametri tecnici. Tra i dati fondamentali vi sono i dataset S&P 500 e i rating creditizi.<sup><a href="#ref-10">[10]</a></sup> Tra i parametri strategici figura il <b>profilo di rischio</b>, che imposta automaticamente i pesi dello score e i limiti di concentrazione. Un profilo conservativo privilegia rating alti e bassa correlazione, mentre uno aggressivo tollera rischi maggiori per rendimenti attesi più alti.<sup><a href="#ref-7">[7]</a></sup></p>
<h3 style="margin-top: 25px; margin-bottom: 10px; color: #34495e;">2.4 Risultati attesi</h3>
<p style="text-align: justify; margin-bottom: 15px;">L’implementazione produce l’universo ridotto $U'$, il grafo filtrato (utile per visualizzare le connessioni) e, soprattutto, due portafogli principali: quello ottimizzato tramite Branch & Bound e quello “decorrelato” tramite Dijkstra.<sup><a href="#ref-3">[3]</a></sup> Per entrambi vengono fornite metriche di rischio/rendimento e le proiezioni future derivanti dalla simulazione Monte Carlo.<sup><a href="#ref-8">[8]</a></sup></p>
<h3 style="margin-top: 25px; margin-bottom: 10px; color: #34495e;">2.5 Criticità, potenzialità e significato</h3>
<p style="text-align: justify; margin-bottom: 15px;">Una criticità centrale è la <b>stima dei parametri</b> (covarianza e rendimenti), sensibile al rumore: lo shrinkage è essenziale ma la scelta di $\lambda$ è delicata.<sup><a href="#ref-1">[1]</a></sup> Anche la costruzione del grafo dipende fortemente dalle soglie $\tau$ e $k$. Tuttavia, il modello offre grandi <b>potenzialità</b> grazie alla flessibilità dei pesi nello score ($\alpha, \beta, \dots$) e alla capacità di integrare vincoli aziendali reali (rating, settore). Combinando ottimizzazione combinatoria, teoria dei grafi e simulazione, il sistema rappresenta un prototipo operativo concreto per la gestione di portafogli vincolati.<sup><a href="#ref-4">[4]</a></sup></p>

<h2 style="margin-top: 30px; margin-bottom: 15px; color: #2c3e50;">3. Descrizione del data-set utilizzato per l’analisi</h2>
<p style="text-align: justify; margin-bottom: 15px;">Il lavoro si basa su due dataset pubblici in formato CSV, integrati in un unico universo di investimento. Il primo è il file “all_stocks_5yr.csv” (Kaggle), che contiene quotazioni giornaliere (Open, High, Low, Close, Volume) per 505 titoli azionari USA su un orizzonte di circa cinque anni, dal 2013-02-08 al 2018-02-07. A partire da questo file sono stati costruiti, per ciascun ticker, una serie storica di prezzi “close” omogenea per calendario e i rendimenti logaritmici giornalieri. Sono stati rimossi i titoli con storia troppo corta o con lacune eccessive, e si è ottenuta una matrice prezzi utilizzata come base per il calcolo di media, volatilità, correlazioni e per la simulazione Monte Carlo.<sup><a href="#ref-10">[10]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Il secondo dataset è “corporateCreditRatingWithFinancialRatios.csv”, che raccoglie 7.805 osservazioni di rating aziendali con informazioni sulle principali metriche di bilancio e metadati quali nome della società, ticker, settore e agenzia di rating. Da questo file si è ricavato, per ogni ticker, un rating alfabetico consolidato, poi mappato in un punteggio numerico su scala discreta (AAA→22, …, D→1). Il processo di data cleaning ha incluso l’eliminazione di osservazioni duplicate e la selezione di un’unica osservazione per ciascun emittente. I due dataset sono stati uniti mediante il ticker, ottenendo un profilo completo che costituisce la base informativa per la costruzione del grafo di correlazione e l’ottimizzazione combinatoria.<sup><a href="#ref-6">[6]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Su questi data-set vengono costruite le grandezze derivate usate dal modello:</p>
<ul style="margin-bottom: 15px; padding-left: 40px;">
<li style="text-align: justify; margin-bottom: 5px;">matrice di correlazione $\rho$ e matrice delle distanze $d_{ij}$ (ad es. $d_{ij} = 1 - |\rho_{ij}|$ o varianti signed);</li>
<li style="text-align: justify; margin-bottom: 5px;">misure sintetiche di rischio (varianza, volatilità, Sharpe ratio) a livello di portafoglio;</li>
<li style="text-align: justify; margin-bottom: 5px;">indicatori sul mix rated/unrated (quota percentuale, rating medio pesato).<sup><a href="#ref-3">[3]</a></sup></li>
</ul>

<h2 style="text-align: center; margin-top: 30px; margin-bottom: 15px; color: #2c3e50;">4. Descrizione ad alto livello delle strutture dati e degli algoritmi utilizzati</h2>
<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">4.1 Strutture dati principali</h3>
<p style="text-align: justify; margin-bottom: 15px;">Il cuore della gestione dati ruota attorno alla classe <code>Stock</code>. Questo oggetto funge da contenitore per tutte le informazioni rilevanti di un singolo titolo: il ticker, il settore di appartenenza, il rating (sia in formato numerico che la data di assegnazione) e, naturalmente, le serie storiche dei prezzi e dei rendimenti.<sup><a href="#ref-2">[2]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">La logica di base è incapsulata nel metodo <code>set_prices</code>. Quando viene fornita una serie di prezzi di chiusura ("close"), la classe si occupa di ordinarla temporalmente, convertire i valori in formato float e calcolare immediatamente i rendimenti logaritmici. Questo passaggio è fondamentale per avere dati pronti all'uso nelle fasi successive di stima del rischio.<sup><a href="#ref-11">[11]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
class Stock:<br>
&nbsp;&nbsp;&nbsp;&nbsp;def set_prices(self, prices: pd.Series) -> None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Ordina la serie per data e si assicura che sia float<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;prices = prices.sort_index().astype(float)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.prices = prices<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Calcola log-prezzi e rendimenti logaritmici<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;log_prices = prices.apply(np.log)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.returns = log_prices.diff()
</div>
<p style="text-align: justify; margin-bottom: 15px;">Per recuperare i dati senza complicare la logica principale, si usa il <code>DAO</code>. Questa classe gestisce la lettura dei file CSV grezzi e costruisce le strutture necessarie: un DataFrame per i prezzi (<code>prices_df</code>), uno per i rating (<code>ratings_df</code>) e infine un dizionario che mappa ogni ticker al suo oggetto <code>Stock</code>.<sup><a href="#ref-10">[10]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
class DAO:<br>
&nbsp;&nbsp;&nbsp;&nbsp;def _load_prices(self) -> pd.DataFrame:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;df = pd.read_csv(self._prices_path)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;df["date"] = pd.to_datetime(df["date"])<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;df = df.sort_values("date")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Pivot: index = date, columns = ticker, values = close<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;prices = df.pivot(index="date", columns="Name", values="close")<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;prices = prices.sort_index().astype(float)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self._prices_df = prices<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return self._prices_df
</div>
<p style="text-align: justify; margin-bottom: 15px;">Parallelamente, <code>_load_ratings</code> gestisce le valutazioni del merito creditizio. Poiché i rating sono spesso espressi in lettere (es. "AAA", "BB+"), viene utilizzata una mappa di conversione per trasformarli in un punteggio numerico (<code>rating_score</code>) su cui è possibile fare calcoli matematici.</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;def _load_ratings(self) -> pd.DataFrame:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;df = pd.read_csv(self._ratings_path)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rating_map = {<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"AAA": 22, "AA+": 21, "AA": 20, "AA-": 19,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"A+": 18,  "A": 17,   "A-": 16,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"BBB+": 15, "BBB": 14, "BBB-": 13,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"BB+": 12,  "BB": 11,  "BB-": 10,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"B+": 9,    "B": 8,    "B-": 7,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"CCC+": 6,  "CCC": 5,  "CCC-": 4,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"CC": 3,    "C": 2,    "D": 1,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;}<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;df["Rating Date"] = pd.to_datetime(df["Rating Date"])<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;df["rating_score"] = df["Rating"].map(rating_map)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self._ratings_df = df<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return self._ratings_df
</div>
<p style="text-align: justify; margin-bottom: 15px;">La funzione <code>build_stock_dict</code> unisce questi due flussi di dati. Per ogni colonna presente nel DataFrame dei prezzi, viene istanziato un oggetto <code>Stock</code> e arricchito con l'ultimo rating disponibile, se presente. Questo dizionario diventa lo stato globale che il <code>Model</code> utilizzerà per tutte le elaborazioni.
</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;def build_stock_dict(self) -> Dict[str, Stock]:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self._prices_df is None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self._load_prices()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self._ratings_df is None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self._load_ratings()<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;stock_dict: Dict[str, Stock] = {}<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for ticker in self._prices_df.columns:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;s = Stock(ticker=ticker)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Serie prezzi per il ticker (può avere NaN)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;prices_series = self._prices_df[ticker].dropna()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if prices_series.empty:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;continue<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;s.set_prices(prices_series)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Rating più recente, se esiste<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;subset = self._ratings_df[self._ratings_df["Ticker"] == ticker]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if not subset.empty:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;latest = subset.sort_values("Rating Date").iloc[-1]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;s.rating_score = latest["rating_score"]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;s.rating_date = latest["Rating Date"]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;s.sector = latest["Sector"]<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;stock_dict[ticker] = s<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self._stock_dict = stock_dict<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return stock_dict
</div>
<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">4.2 Stima del rischio: rendimenti, correlazioni e covarianza shrinkata</h3>
<p style="text-align: justify; margin-bottom: 15px;">Una volta ottenuti i dati grezzi, la classe <code>RiskEstimator</code> si occupa di trasformarli in metriche statistiche utilizzabili. Il processo segue una pipeline precisa: calcolo dei rendimenti, pulizia degli outlier (winsorization), filtraggio dei dati mancanti e stima finale dei parametri di rischio (media $\mu$, covarianza $\Sigma$ e correlazione $\rho$).<sup><a href="#ref-2">[2]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Il primo passo è semplice: si calcola la variazione logaritmica dei prezzi giornalieri.</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
class RiskEstimator:<br>
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def compute_returns(prices_df: pd.DataFrame) -> pd.DataFrame:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;prices = prices_df.sort_index().astype(float)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# R_t = log(P_t / P_{t-1})<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;returns = np.log(prices / prices.shift(1))<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return returns
</div>
<p style="text-align: justify; margin-bottom: 15px;">I dati finanziari reali presentano spesso valori anomali che possono distorcere le stime. Per mitigare questo problema, il metodo <code>winsorize</code> limita i valori estremi entro specifici quantili (di default l'1% e il 99%), "schiacciando" gli outlier verso i bordi della distribuzione.

 <sup><a href="#ref-11">[11]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def winsorize(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;returns_df: pd.DataFrame,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;lower: float = 0.01,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;upper: float = 0.99<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> pd.DataFrame:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if returns_df is None or returns_df.empty:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;raise ValueError("returns_df è vuoto.")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;q_low = returns_df.quantile(lower)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;q_high = returns_df.quantile(upper)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Clamp colonna per colonna<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;wins = returns_df.clip(lower=q_low, upper=q_high, axis=1)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return wins
</div>
<p style="text-align: justify; margin-bottom: 15px;">La funzione principale <code>estimate_from_returns</code> orchestra tutto il processo. Oltre alla winsorization, si applica un filtro per escludere i titoli con troppi dati mancanti (<code>min_non_na_ratio</code>) e si imputano i restanti NaN con la media. Successivamente, calcola le matrici standard di covarianza e correlazione.</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def estimate_from_returns(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;returns_df: pd.DataFrame,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;shrink_lambda: float,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;min_non_na_ratio: float,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;winsor_lower: Optional[float],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;winsor_upper: Optional[float],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;shrink_target: str = "diag"<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> Tuple[pd.DataFrame, pd.DataFrame, pd.Series]:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 1) Winsorization opzionale<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if winsor_lower is not None and winsor_upper is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;returns_df = RiskEstimator.winsorize(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;returns_df,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;lower=winsor_lower,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;upper=winsor_upper,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 2) Elimina righe completamente NaN<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;returns_df = returns_df.dropna(how="all")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 3) Tiene solo colonne con abbastanza dati non-NaN<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;non_na_ratio = returns_df.notna().mean(axis=0)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mask_ok = non_na_ratio >= min_non_na_ratio<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;returns_df = returns_df.loc[:, mask_ok]<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 4) Imputazione NaN rimanenti con la media di colonna<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;returns_df = returns_df.fillna(returns_df.mean())<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 5) Stima μ, Σ e ρ<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mu = returns_df.mean(axis=0)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Sigma = returns_df.cov()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rho = returns_df.corr()
</div>
<p style="text-align: justify; margin-bottom: 15px;">Per rendere la matrice di covarianza più robusta (specialmente quando il numero di osservazioni non è elevatissimo rispetto al numero di titoli), si applica uno <i>shrinkage</i> lineare verso una matrice diagonale. In pratica, si fa una media pesata tra la matrice campionaria e una matrice ideale che contiene solo le varianze sulla diagonale.<sup><a href="#ref-1">[1]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Target diagonale: solo varianze sulla diagonale<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if shrink_target == "diag":<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;T = pd.DataFrame(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;np.diag(np.diag(Sigma.values)),<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;index=Sigma.index,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;columns=Sigma.columns,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;raise ValueError(f"Target di shrinkage non supportato: {shrink_target}")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Shrinkage elementare: Sigma_sh = (1 - λ) Σ + λ T<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Sigma_sh = (1.0 - shrink_lambda) * Sigma + shrink_lambda * T<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return rho, Sigma_sh, mu
</div>
<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">4.3 Costruzione del grafo di correlazione e universo ridotto U′</h3>
<p style="text-align: justify; margin-bottom: 15px;">Per applicare algoritmi basati sulla teoria dei grafi, la classe <code>GraphBuilder</code> converte la matrice di correlazione <code>rho</code> in una matrice di distanze. La logica è che due titoli molto correlati sono "vicini", mentre titoli decorrelati sono "lontani".<sup><a href="#ref-4">[4]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
class GraphBuilder:<br>
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def build_distance_matrix(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rho: pd.DataFrame,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;signed: bool = False<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> pd.DataFrame:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if rho is None or rho.empty:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;raise ValueError("rho è vuota.")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if signed:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# d_ij = (1 - ρ_ij) / 2  ∈ [0,1]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;d_values = (1.0 - rho.values) / 2.0<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# d_ij = 1 - |ρ_ij|  ∈ [0,1]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;d_values = 1.0 - np.abs(rho.values)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;d = pd.DataFrame(d_values, index=rho.index, columns=rho.columns)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;np.fill_diagonal(d.values, 0.0)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return d
</div>
<p style="text-align: justify; margin-bottom: 15px;">Per costruire il grafo vero e proprio, spesso è utile filtrare il rumore rimuovendo le connessioni deboli. Si usano due metodi: un filtro a soglia fissa (che elimina archi con correlazione troppo bassa) e un filtro k-NN (che mantiene solo i <i>k</i> vicini più stretti per ogni nodo).
 <sup><a href="#ref-5">[5]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def threshold_filter(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rho: pd.DataFrame,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;tau: float<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> pd.DataFrame:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if rho is None or rho.empty:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;raise ValueError("rho è vuota.")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;abs_rho = np.abs(rho)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Distanze: 1 - |ρ_ij|, ma se |ρ_ij| < τ → +inf (nessun arco)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist = 1.0 - abs_rho<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mask_below = abs_rho.values < tau<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist.values[mask_below] = np.inf<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;np.fill_diagonal(dist.values, 0.0)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return dist
</div>
<p style="text-align: justify; margin-bottom: 15px;">Il filtro k-NN è leggermente più complesso perché deve garantire la simmetria del grafo finale (se A è vicino a B, vogliamo che l'arco esista in entrambe le direzioni).<sup><a href="#ref-3">[3]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def knn_filter(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;distance: pd.DataFrame,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;k: int,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;symmetric: bool = True<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> pd.DataFrame:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if distance is None or distance.empty:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;raise ValueError("distance è vuota.")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist = distance.copy()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;n = dist.shape[0]<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for i in range(n):<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;row = dist.iloc[i, :].values<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# indici delle k distanze più piccole (escludendo i=i)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;idx_sorted = np.argsort(row)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;keep = idx_sorted[:k+1]   # include la diagonale<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mask_drop = np.ones_like(row, dtype=bool)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mask_drop[keep] = False<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;row[mask_drop] = np.inf<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist.iloc[i, :] = row<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if symmetric:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist_vals = dist.values<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist_sym = np.minimum(dist_vals, dist_vals.T)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist = pd.DataFrame(dist_sym, index=dist.index, columns=dist.columns)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;np.fill_diagonal(dist.values, 0.0)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return dist
</div>
<p style="text-align: justify; margin-bottom: 15px;">Combinando questi elementi, <code>build_filtered_graph</code> restituisce un oggetto grafo di NetworkX, pronto per essere analizzato. Il modello utilizza queste informazioni anche per definire un "universo ridotto" $U'$, selezionando preliminarmente i titoli meno correlati con il resto del mercato e rispettando i vincoli di rating.<sup><a href="#ref-4">[4]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def build_filtered_graph(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rho: pd.DataFrame,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;tau: Optional[float] = None,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;k: Optional[int] = None,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;signed: bool = False<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> Tuple[nx.Graph, pd.DataFrame, pd.DataFrame]:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;abs_rho = np.abs(rho)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 1) Matrice delle distanze base<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist = GraphBuilder.build_distance_matrix(rho, signed=signed)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 2) Filtro a soglia (se tau è fornito)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if tau is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mask_below = abs_rho.values < tau<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist.values[mask_below] = np.inf<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 3) Filtro k-NN (se k è fornito)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if k is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist_knn = GraphBuilder.knn_filter(dist, k=k, symmetric=True)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dist_knn = dist<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;adj = abs_rho.copy()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mask_no_edge = ~np.isfinite(dist_knn.values)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;adj.values[mask_no_edge] = 0.0<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Crea grafo NetworkX con attributi<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;G = nx.Graph()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for i, u in enumerate(adj.index):<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for j, v in enumerate(adj.columns):<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w = adj.iat[i, j]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if i < j and w > 0.0:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;G.add_edge(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;u, v,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;weight=w,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;corr=rho.iat[i, j],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;distance=dist_knn.iat[i, j],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return G, adj, dist_knn
</div>
<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">4.4 Selezione combinatoria del portafoglio: pre-filtro e Branch & Bound</h3>
<p style="text-align: justify; margin-bottom: 15px;">La classe <code>PortfolioSelector</code> gestisce la selezione ottimale dei titoli. Dato che un'esplorazione esaustiva su centinaia di titoli è computazionalmente impossibile, si procede in due fasi. La prima è un <b>pre-filtro</b> (<code>build_selector_universe</code>) che calcola uno score individuale per ogni asset basato su rendimento, volatilità e rating.<sup><a href="#ref-6">[6]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
class PortfolioSelector:<br>
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def build_selector_universe(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;base_universe: List[str],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;K: int,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mu: pd.Series,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Sigma_sh: pd.DataFrame,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;stocks: Dict[str, Stock],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;a1: float,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;a2: float,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;a3: float,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;min_size: int = 10,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;factor: int = 4,<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> List[str]:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;candidates = []<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for t in base_universe:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if t not in mu.index or t not in Sigma_sh.index:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;continue<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mu_i = float(mu.loc[t])<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;sigma_i = float(np.sqrt(Sigma_sh.loc[t, t]))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;stock = stocks.get(t)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rating_score = stock.rating_score if stock else None<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rating_norm = 0.0<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if rating_score is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rating_norm = (rating_score - 1) / (22 - 1)  # normalizza in [0,1]<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# asset_score = a1 * μ_i - a2 * σ_i + a3 * rating_norm<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;asset_score = a1 * mu_i - a2 * sigma_i + a3 * rating_norm<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;candidates.append((t, asset_score))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Ordina per asset_score decrescente<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;candidates.sort(key=lambda x: x[1], reverse=True)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Prende i primi max(factor*K, min_size)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;limit = max(factor * K, min_size)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;selected = [t for t, _ in candidates[:limit]]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return selected
</div>
<p style="text-align: justify; margin-bottom: 15px;">Sull'insieme ristretto di candidati, si applica l'algoritmo vero e proprio. Per efficienza, i dati vengono convertiti in matrici NumPy. La funzione <code>_violates_constraints_indices</code> controlla i vincoli "hard" (es. rating minimo, concentrazione settoriale, massima correlazione a coppie) ogni volta che si prova ad aggiungere un titolo.<sup><a href="#ref-7">[7]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;def _violates_constraints_indices(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;indices: List[int],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;new_idx: int<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> bool:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 1) Aggiunge provvisoriamente il nuovo indice<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;cand = indices + [new_idx]<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 2) Rating minimo: i titoli rated devono avere rating >= rating_min<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self.rating_min is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ratings = self.rating_scores_np[cand]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mask_rated = ~np.isnan(ratings)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if mask_rated.any():<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if (ratings[mask_rated] < self.rating_min).any():<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return True<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 3) Quota massima di unrated<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self.max_unrated_share is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ratings = self.rating_scores_np[cand]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;n_unrated = np.isnan(ratings).sum()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;share_unrated = n_unrated / len(cand)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if share_unrated > self.max_unrated_share:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return True<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 4) Concentrazione settoriale<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self.max_sector_size is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;sectors = self.sector_ids_np[cand]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;_, counts = np.unique(sectors, return_counts=True)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if (counts > self.max_sector_size).any():<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return True<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 5) Correlazione massima per coppia<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self.rho_pair_max is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for i in cand:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for j in cand:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if i < j:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if np.abs(self.rho_np[i, j]) > self.rho_pair_max:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return True<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return False
</div>
<p style="text-align: justify; margin-bottom: 15px;">La funzione obiettivo (<code>_score_indices</code>) valuta la qualità di un portafoglio candidato combinando decorrelazione, rating medio e rendimento atteso, penalizzando eventuali eccessi di concentrazione settoriale.</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;def _score_indices(self, indices: List[int]) -> float:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if len(indices) < 2:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return -np.inf<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 1) Correlazione media assoluta<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;sub_rho = self.rho_np[np.ix_(indices, indices)]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;vals = sub_rho[np.triu_indices(len(indices), k=1)]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mean_corr = float(np.mean(np.abs(vals)))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 2) Rating medio (solo rated)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;r_vals = self.rating_scores_np[indices]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mask_rated = ~np.isnan(r_vals)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mean_rating = float(np.mean(r_vals[mask_rated])) if mask_rated.any() else 0.0<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 3) Penalità settoriale<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;sectors = self.sector_ids_np[indices]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;_, counts = np.unique(sectors, return_counts=True)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;excess = np.clip(counts - self.max_sector_size, a_min=0, a_max=None)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;pen_sector = float(np.sum(excess))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# 4) Rendimento medio<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mu_vals = self.mu_np[indices]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mean_return = float(np.mean(mu_vals))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;score = (<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.alpha * (-mean_corr)   # decorrelazione<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;+ self.beta  * mean_rating  # qualità del credito<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- self.gamma * pen_sector   # penalità settoriale<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;+ self.delta * mean_return  # rendimento<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return score
</div>
<p style="text-align: justify; margin-bottom: 15px;">L'algoritmo di ricerca è un <b>Branch & Bound</b>. Esplora l'albero delle combinazioni in profondità, ma utilizza un <i>bound</i> ottimistico per "potare" i rami inutili. Se la somma parziale dei rendimenti non è sufficiente a superare il miglior punteggio trovato finora, quel ramo viene abbandonato.<sup><a href="#ref-6">[6]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;def _search(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;partial_indices: List[int],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;start_pos: int,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;sum_mu_partial: float<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;k_partial = len(partial_indices)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Se abbiamo già K titoli, valuta lo score<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if k_partial == self.K:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;score = self._score_indices(partial_indices)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if score > self.best_score:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.best_score = score<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.best_indices = partial_indices.copy()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Bound ottimistico sul contributo di μ<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;max_extra = 0.0<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;remaining = self.K - k_partial<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if remaining > 0:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;extra_mu = self.mu_sorted[start_pos : start_pos + remaining]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;max_extra = float(np.sum(extra_mu))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Se anche nel caso migliore non miglioriamo il best_score, prune<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;bound_mu = sum_mu_partial + max_extra<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if bound_mu + self.upper_bound_other_terms <= self.best_score:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for pos in range(start_pos, len(self.sorted_indices)):<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;idx = self.sorted_indices[pos]<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self._violates_constraints_indices(partial_indices, idx):<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;continue<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;new_indices = partial_indices + [idx]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;new_sum_mu = sum_mu_partial + self.mu_np[idx]<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self._search(new_indices, pos + 1, new_sum_mu)
</div>
<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">4.5 Calcolo dei pesi di portafoglio</h3>
<p style="text-align: justify; margin-bottom: 15px;">Una volta che l'algoritmo combinatorio ha identificato <i>quali</i> titoli comprare, la classe <code>PortfolioWeights</code> decide <i>quanto</i> comprarne. Si supportano due modalità: pesi uguali (Equal Weight) o ottimizzazione Media-Varianza. Nel caso Media-Varianza, si risolve il classico problema di minimizzazione della varianza dato un livello di avversione al rischio, aggiungendo un termine di regolarizzazione (ridge) per stabilità numerica.  <sup><a href="#ref-2">[2]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
class PortfolioWeights:<br>
&nbsp;&nbsp;&nbsp;&nbsp;@staticmethod<br>
&nbsp;&nbsp;&nbsp;&nbsp;def mean_variance_weights(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;tickers: List[str],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mu: pd.Series,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Sigma_sh: pd.DataFrame,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;risk_aversion: float = 1.0,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ridge: float = 1e-4,<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> pd.Series:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;n = len(tickers)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mu_vec = mu.loc[tickers].values.reshape(-1, 1)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Sigma_sub = Sigma_sh.loc[tickers, tickers].values<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Regolarizza la covarianza (evita problemi di condizionamento)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Sigma_reg = Sigma_sub + ridge * np.eye(n)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A = np.linalg.inv(Sigma_reg)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ones = np.ones((n, 1))<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;c = risk_aversion * mu_vec<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Formula chiusa per il problema quadratico con vincolo 1ᵀw = 1<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;lam = (ones.T @ (A @ c) - 1.0) / (ones.T @ (A @ ones))<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_raw = A @ (c - lam * ones)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_raw = w_raw.flatten()
</div>
<p style="text-align: justify; margin-bottom: 15px;">Poiché si tratta di portafogli "long-only", i pesi negativi vengono azzerati e si applica un vincolo di peso minimo per evitare posizioni troppo piccole. Infine, si normalizza e arrotonda per garantire che la somma sia esattamente 1.<sup><a href="#ref-6">[6]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Long-only<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_raw = np.maximum(w_raw, 0.0)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Peso minimo w_min = 1/(2K)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;K = n<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_min = 1.0 / (2.0 * K)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;base = np.full(K, w_min)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;residual = 1.0 - np.sum(base)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if residual < 0:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# fallback conservativo<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w = np.full(K, 1.0 / K)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if w_raw.sum() > 0:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_extra = w_raw / w_raw.sum()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_extra = np.full(K, 1.0 / K)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w = base + residual * w_extra<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Arrotonda a due decimali mantenendo somma a 1<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_rounded = np.round(w, 2)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;diff = 1.0 - w_rounded.sum()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if np.abs(diff) > 1e-6:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;idx_max = np.argmax(w_rounded)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_rounded[idx_max] += diff<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return pd.Series(w_rounded, index=tickers)
</div>
<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">4.6 Portafoglio min–correlazione via Dijkstra</h3>
<p style="text-align: justify; margin-bottom: 15px;">In alternativa all'approccio combinatorio puro, il sistema implementa una strategia euristica basata sulla geometria del grafo. L'idea è scegliere un titolo "sorgente" e selezionare gli altri titoli massimizzando la distanza (e quindi minimizzando la correlazione) da esso.

 <sup><a href="#ref-9">[9]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Il metodo <code>build_graph_for_dijkstra</code> prepara il grafo filtrato:</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
class Model:<br>
&nbsp;&nbsp;&nbsp;&nbsp;def build_graph_for_dijkstra(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;use_reduced_universe: bool = True,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;signed: bool = False,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;tau: Optional[float] = None,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;k: Optional[int] = None,<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self.current_rho is None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;raise RuntimeError("rho non stimata.")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if use_reduced_universe and self.reduced_universe is not None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rho_sub = self.current_rho.loc[self.reduced_universe, self.reduced_universe]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rho_sub = self.current_rho<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;G, adj, dist = GraphBuilder.build_filtered_graph(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rho_sub,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;tau=tau,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;k=k,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;signed=signed,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.current_graph = G<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.current_adj = adj<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.current_dist = dist
</div>
<p style="text-align: justify; margin-bottom: 15px;">Si calcolano poi i percorsi minimi dal nodo sorgente verso tutti gli altri nodi. I titoli vengono ordinati in base alla loro distanza dal sorgente: quelli più "lontani" sono teoricamente i migliori diversificatori rispetto al punto di partenza.<sup><a href="#ref-4">[4]</a></sup></p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;def _compute_dijkstra_distances_from_source(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;source: str<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> Dict[str, float]:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self.current_graph is None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;raise RuntimeError("Grafo non costruito.")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Distanze minime dal nodo source<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;lengths = nx.single_source_dijkstra_path_length(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self.current_graph,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;source,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;weight="distance",<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return lengths
</div>
<p style="text-align: justify; margin-bottom: 15px;">Durante la costruzione del portafoglio, si scorre la lista dei candidati e si aggiungono al portafoglio solo se rispettano il vincolo sulla correlazione massima a coppie (<code>_ok_rho_pairwise</code>).</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;def build_portfolio_dijkstra(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;source: str,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;K: int,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;use_reduced_universe: bool = True,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;require_rating: bool = False,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;min_rating_score: float = 13.0,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rho_pair_max: Optional[float] = None,<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> List[str]:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;universe = self._build_dijkstra_universe(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;source, K, use_reduced_universe, require_rating, min_rating_score<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;lengths = self._compute_dijkstra_distances_from_source(source)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;candidates_dist = [<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(t, d) for t, d in lengths.items()<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if t != source and t in universe<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# ordina per distanza decrescente<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;candidates_dist.sort(key=lambda x: x[1], reverse=True)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;portfolio = [source]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for t, d in candidates_dist:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if len(portfolio) >= K:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;break<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self._ok_rho_pairwise(portfolio, t, rho_pair_max):<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;portfolio.append(t)
</div>
<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">4.7 Simulazioni Monte Carlo e orchestrazione applicativa</h3>
<p style="text-align: justify; margin-bottom: 15px;">Per valutare la robustezza dei portafogli generati, il sistema esegue simulazioni Monte Carlo. Invece di affidarsi solo ai dati storici passati, si generano migliaia di scenari futuri possibili basati sui parametri $\mu$ e $\Sigma$ stimati.  <sup><a href="#ref-8">[8]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">La funzione <code>simulate_portfolio_paths</code> genera traiettorie di prezzo assumendo che i rendimenti giornalieri seguano una distribuzione normale multivariata.</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
class Model:<br>
&nbsp;&nbsp;&nbsp;&nbsp;def simulate_portfolio_paths(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;tickers: List[str],<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;weights: pd.Series,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;n_paths: int = 100,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;n_days: int = 252,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;seed: Optional[int] = None,<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> np.ndarray:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if self.current_mu is None or self.current_Sigma_sh is None:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;raise RuntimeError("Parametri di rischio non stimati.")<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_vec = weights.loc[tickers].values.astype(float)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w_vec = w_vec / w_vec.sum()<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mu_vec = self.current_mu.loc[tickers].values<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Sigma_sub = self.current_Sigma_sh.loc[tickers, tickers].values<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mu_p = float(w_vec @ mu_vec)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;sigma_p = float(np.sqrt(w_vec @ Sigma_sub @ w_vec))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;rng = np.random.default_rng(seed)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Genera shock giornalieri ~ N(μ_p, σ_p)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;shocks = rng.normal(loc=mu_p, scale=sigma_p, size=(n_paths, n_days))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Somma cumulata dei log-rendimenti<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;log_paths = np.cumsum(shocks, axis=1)<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;# Valore iniziale 1, poi applica exp<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;paths = np.exp(np.hstack([np.zeros((n_paths, 1)), log_paths]))<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return paths
</div>
<p style="text-align: justify; margin-bottom: 15px;">Infine, <code>_summarize_mc_paths</code> condensa queste simulazioni in statistiche chiave: rendimento medio atteso, volatilità, VaR (tramite i percentili) e probabilità di perdita.</p>
<div style="text-align: left; background-color: #f6f8fa; padding: 15px; border-radius: 6px; border: 1px solid #d0d7de; font-family: 'Courier New', monospace; font-size: 0.9em; overflow-x: auto; margin-bottom: 20px;">
&nbsp;&nbsp;&nbsp;&nbsp;def _summarize_mc_paths(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;self,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;paths: np.ndarray<br>
&nbsp;&nbsp;&nbsp;&nbsp;) -> Dict[str, float]:<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;final_vals = paths[:, -1]<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;mean_val = float(final_vals.mean())<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;std_val = float(final_vals.std())<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;q05, q50, q95 = np.percentile(final_vals, [5, 50, 95])<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;prob_loss = float(np.mean(final_vals < 1.0))<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return {<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"mean": mean_val,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"std": std_val,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"q05": float(q05),<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"q50": float(q50),<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"q95": float(q95),<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"prob_loss": prob_loss,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;}
</div>

<h2 style="text-align: center; margin-top: 30px; margin-bottom: 15px; color: #2c3e50;">5. Valutazioni sui risultati ottenuti e conclusioni</h2>
<p style="text-align: justify; margin-bottom: 15px;">L’ultima parte del lavoro è dedicata alla verifica pratica del prototipo sviluppato. <h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">5.1 Analisi dei tempi di convergenza (convergence plots)</h3>
<p style="text-align: justify; margin-bottom: 15px;">Quando il portafoglio è molto piccolo ($K = 3$), l’algoritmo risulta rapido. Su universi già filtrati, composti da una trentina di titoli circa, il tempo di esecuzione si colloca sotto il secondo. Anche con un universo leggermente più ampio si resta comunque intorno a 1–2 secondi, grazie al pre-filtro che scarta rapidamente le soluzioni peggiori.<sup><a href="#ref-6">[6]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Già con $K = 4$ il quadro cambia. I tempi iniziano a crescere in modo evidente: per il profilo prudente si osservano mediamente 8–10 secondi, mentre per il profilo bilanciato i tempi possono superare i 20 secondi. L’effetto combinatorio diventa molto più pesante nonostante la potatura effettuata dal Branch & Bound.</p>
<p style="text-align: justify; margin-bottom: 15px;">Con $K = 5$ e $K = 6$ la crescita dei tempi assume di fatto un comportamento quasi esplosivo. Se l’universo iniziale si allarga, i tempi possono aumentare di un ordine di grandezza (es. da 40 a 380 secondi). È un segnale chiaro della natura combinatoria del problema: anche con un buon pruning, il numero di combinazioni potenzialmente valide resta molto elevato.  <sup><a href="#ref-6">[6]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">A fronte di queste difficoltà, l’algoritmo basato su Dijkstra si comporta in modo opposto. Una volta costruito il grafo filtrato, il calcolo dei percorsi minimi richiede tempi trascurabili (inferiori a 0,1 secondi), poiché il costo computazionale è concentrato nella costruzione del grafo eseguita una tantum.<sup><a href="#ref-9">[9]</a></sup></p>

<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">5.3 Punti di debolezza e limiti</h3>
<ul style="margin-bottom: 15px; padding-left: 40px;">
<li style="text-align: justify; margin-bottom: 5px;"><b>Qualità dei dati:</b> Le stime derivano da dati storici e sono affette da rumore. Lo shrinkage aiuta, ma non elimina l'incertezza sui rendimenti futuri.<sup><a href="#ref-2">[2]</a></sup></li>
<li style="text-align: justify; margin-bottom: 5px;"><b>Parametri di grafo:</b> La scelta di $\tau$ e $k$ impatta pesantemente sulla struttura della rete. Valori troppo severi spezzano il grafo, valori permissivi aumentano il rumore.<sup><a href="#ref-5">[5]</a></sup></li>
<li style="text-align: justify; margin-bottom: 5px;"><b>Scalabilità:</b> Il Branch & Bound fatica a scalare su universi ampi. In scenari reali sarebbe opportuno affiancare euristiche per soluzioni approssimate più rapide.<sup><a href="#ref-6">[6]</a></sup></li>
<li style="text-align: justify; margin-bottom: 5px;"><b>Ipotesi Monte Carlo:</b> L'assunzione di normalità dei rendimenti è una semplificazione che ignora le code grasse e gli shock sistemici tipici dei mercati reali.<sup><a href="#ref-8">[8]</a></sup></li>
<li style="text-align: justify; margin-bottom: 5px;"><b>Staticità:</b> Il portafoglio è costruito "una tantum" senza ribilanciamenti o costi di transazione, elementi che nella pratica modificherebbero il profilo netto.<sup><a href="#ref-4">[4]</a></sup></li>
</ul>
<h3 style="text-align: center; margin-top: 25px; margin-bottom: 10px; color: #34495e;">5.4 Conclusioni e risposta al problema aziendale</h3>
<p style="text-align: justify; margin-bottom: 15px;">Il lavoro offre una risposta concreta all'esigenza di costruire portafogli vincolati (rating, settore, correlazione). Il prototipo dimostra che il filtraggio su grafo riduce efficacemente il rumore e che l'algoritmo Branch & Bound, pur con limiti di scalabilità, gestisce vincoli complessi altrimenti intrattabili.<sup><a href="#ref-3">[3]</a></sup></p>
<p style="text-align: justify; margin-bottom: 15px;">Dal confronto emerge che non esiste un vincitore assoluto: nel profilo prudente le strategie si equivalgono; nel profilo bilanciato l'approccio topologico (Dijkstra) offre più crescita (e rischio); nel profilo dinamico l'ottimizzazione combinatoria vince grazie alla selezione di titoli ad alto drift. Il prototipo è quindi uno strumento flessibile, estendibile in futuro con dati real-time e misure di rischio più avanzate.<sup><a href="#ref-7">[7]</a></sup></p>
